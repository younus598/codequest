```python
 import matplotlib.pyplot 
import time
import random
MAX_BATTERY_CAPACITY = int(input("enter the MAX_BATTERY_CAPACITY value"))
FULL_BATTERY_HEALTH = int(input("enter the FULL_BATTERY_HEALTH :"))
MAX_VEHICLE_RANGE = int(input("enter the MAX_VEHICLE_RANGE value:"))
current_battery_health = FULL_BATTERY_HEALTH
current_battery_level = MAX_BATTERY_CAPACITY 
energy_consumption_rate = int(input("enter the energy consumption"))  
def estimate_range(battery_level, consumption_rate):
    return battery_level / consumption_rate
def drive(kilometers):
    global current_battery_level, current_battery_health, energy_consumption_rate
    energy_consumed = kilometers * energy_consumption_rate * random.uniform(0.8, 1.2)
    current_battery_level = max(0, current_battery_level - energy_consumed)
    current_battery_health = max(50, current_battery_health - (kilometers / 1000))
    print(f"Driven {kilometers} km. Energy consumed: {energy_consumed:.2f} kWh.")
    print(f"Remaining battery level: {current_battery_level:.2f} kWh ({(current_battery_level/MAX_BATTERY_CAPACITY) * 100:.2f}%).")
    print(f"Battery health: {current_battery_health:.2f}%.")
def display_energy_dashboard():
    print("\n--- Energy Dashboard ---")
    remaining_range = estimate_range(current_battery_level, energy_consumption_rate)
    print(f"Battery Level: {current_battery_level:.2f} kWh ({(current_battery_level / MAX_BATTERY_CAPACITY) * 100:.2f}%)")
    print(f"Estimated Range: {remaining_range:.2f} km")
    print(f"Battery Health: {current_battery_health:.2f}%")
    print(f"Energy Consumption Rate: {energy_consumption_rate:.2f} kWh/km")
    print("------------------------\n")
def suggest_energy_optimization():
    f = int(input("Enter the con"))
    if current_battery_health < f:
        print("Tip: Your battery health is below 80%. Consider reducing fast charging cycles to prolong battery life.")
    if energy_consumption_rate > 0.35:
        print("Tip: High energy consumption detected. Try smoother acceleration to reduce energy usage.")
    if estimate_range(current_battery_level, energy_consumption_rate) < 50:
        print("Tip: Your remaining range is low. Consider finding the nearest charging station.")
if __name__ == "__main__":
    for i in range(5): 
        kilometers_driven = random.randint(10, 50)  
        drive(kilometers_driven)
        display_energy_dashboard()
        suggest_energy_optimization()
        time.sleep(1)  
```

```python
class ElectricVehicle:
    def __init__(self, battery_health=100, energy_consumption_rate=0, total_distance=0):
        self.battery_health = battery_health  # percentage
        self.energy_consumption_rate = energy_consumption_rate  # kWh/km
        self.total_distance = total_distance  # km
        self.regenerative_braking_efficiency = 0.2  # 20%

    def update_energy_consumption(self, distance):
        energy_used = distance * self.energy_consumption_rate
        self.total_distance += distance
        self.battery_health -= energy_used / 100  # Decrease battery health based on energy used

    def evaluate_performance(self):
        if self.battery_health <= 0:
            self.battery_health = 0
            print("Battery depleted! Please recharge.")
            return
        range_estimate = self.battery_health * 3  # Simplified estimation
        print(f"Estimated Range: {range_estimate:.2f} km")
        print(f"Battery Health: {self.battery_health:.2f}%")
        print(f"Total Distance Driven: {self.total_distance:.2f} km")
        print(f"Regenerative Braking Efficiency: {self.regenerative_braking_efficiency * 100:.2f}%")

    def optimization_suggestions(self):
        suggestions = [
            "Avoid aggressive acceleration.",
            "Use regenerative braking effectively.",
            "Plan your route to avoid heavy traffic.",
            "Charge during off-peak hours."
        ]
        return suggestions


def simulate_driving(ev):
    while True:
        distance = random.uniform(0.5, 5.0)  # Simulate a random distance driven
        ev.update_energy_consumption(distance)
        ev.evaluate_performance()

        # Display suggestions
        print("\nOptimization Suggestions:")
        for suggestion in ev.optimization_suggestions():
            print(f"- {suggestion}")

        print("\n" + "="*30 + "\n")
        time.sleep(5)  # Simulate time between driving sessions


if __name__ == "__main__":
    # Create an Electric Vehicle instance
    ev = ElectricVehicle(energy_consumption_rate=0.2)  # 0.2 kWh/km

    # Start simulating the driving and monitoring process
    simulate_driving(ev)  
```
